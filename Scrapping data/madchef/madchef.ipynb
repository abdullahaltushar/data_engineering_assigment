```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
```

```python
data['Branch_Name']= data['Branch_Name'].apply(lambda x:remove_punctuation(x))
data.head()
```

```python
data['Addres']= data['Addres'].apply(lambda x:remove_punctuation(x))
data.head()
```

```python
import re
def remove_new_line(text):
    line = text.replace('\n','')
    return line
```
